1) removing the null values

```python
frame_dropna = amazon_df.dropna()
```

```python
frame_dropna
```

2) seperating just the product name from cloumn Product_name

```python
def product(element):
    name =''
    products = element.split()
    name = products[0]
    return name
    
    
```

```python
frame_dropna['product'] = frame_dropna['Product_name'].apply(product)
```

```python
frame_dropna.head()
```

3) seperating the Rating column value into the float value

```python
def rate(value):
    star = value.split()
    result = star[0]
    return float(result)
```

```python
frame_dropna['Rating value'] = frame_dropna['Rating'].apply(rate)
frame_dropna.head()
```

5) checking the high rated to low rated laptops.

```python
def checking(rates):
    convt = float(rates)
    if convt == 5:
        return 'rated 5/5'
    elif convt >= 4:
        return 'high rated'
    elif convt < 4 and rates >= 3:
        return 'moderate rated'
    else:
        return 'Low rated'
```

```python
frame_dropna['check_list'] = frame_dropna['Rating value'].apply(checking)
```

```python
frame_dropna['check_list'].value_counts()
```

6) plot to describe the company product rated accordingly

```python
sns.catplot(x='product' ,y='check_list' , hue='product' ,height=5 ,aspect=3, data=frame_dropna )
```
